```python
df = pd.DataFrame()
for current_path, dirs, files in os.walk(path):
  for file in files:
    data_path = f'{current_path}/{file}'
    temp_df = pd.read_csv(data_path, encoding = 'windows-1251', on_bad_lines='skip')
    current_path.split()
    path_parts = current_path.split('/')
    date = path_parts[-2]
    name = path_parts[-1] 
    temp_df['date'] = date
    temp_df['name'] = name
    df = pd.concat((df, temp_df))
    print(df.shape)
```
