```python
taxi = pd.read_csv('/content/taxi_peru.csv', sep = ';')
#taxi = taxi[['start_at', 'end_at', 'arrived_at']].apply(pd.to_datetime)
taxi['start_at'] = pd.to_datetime(taxi.start_at)
taxi['end_at'] = pd.to_datetime(taxi.end_at)
taxi['arrived_at'] = pd.to_datetime(taxi.arrived_at)
taxi.start_type.unique()

```

```python
taxi = taxi.query('start_type == "asap" or start_type == "reserved" ')
```

```python
taxi
```

```python
taxi['minutes'] = (taxi.arrived_at - taxi.start_at).astype('timedelta64[m]')
taxi.dtypes
```

```python
taxi.query('start_type == "reserved" and minutes > 0')\
    .groupby('driver_id')\
    .agg({'journey_id':'count'})\
    .value_counts()
    
```

```python
taxi.query('start_type == "reserved" and minutes > 0')\
    .driver_id\
    .value_counts()
```

```python
taxi['month'] = taxi.start_at.dt.month
taxi['week_day'] = taxi.start_at.dt.strftime('%A') 
taxi
```
