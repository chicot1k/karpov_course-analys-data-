```python
import pandas as pd
```

```python
retail = pd.read_csv('/content/data.csv.zip', encoding = 'ISO-8859-1', compression='zip')
list(retail)
pd.DataFrame(retail)
retail_columns = retail.columns.values.tolist()
retail
```

```python
retail.query('Quantity > 0')
```

```python
retail.quantile(.80)
```

```python
german = retail.query('Country == "Germany"')\
      .groupby('CustomerID', as_index = 0)\
      .agg({'InvoiceNo': pd.Series.nunique})
german
```

```python
german.InvoiceNo.quantile(.80)
```

```python
top_german=german.query('InvoiceNo > 9').CustomerID
```

```python
top_retail_germany=retail.query('CustomerID in @top_german')
top_retail_germany
```

```python
top_retail_germany.query('Description != "POSTAGE"')\
                  .groupby('StockCode')\
                  .agg({'Description':'count'})
```

```python
retail['final'] = retail.Quantity*retail.UnitPrice
retail
```

```python
retail.groupby('InvoiceNo')\
      .agg({'final':'sum'})
```
